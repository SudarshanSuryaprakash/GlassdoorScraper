```python
df_cat = df[['Location', 'Headquarters', 'Size','Type of ownership', 'Industry', 'Sector', 'Revenue','python', 'spark', 'aws', 'excel', 'seniority']]
```

```python
for i in df_cat.columns:
    cat_num = df_cat[i].value_counts()
    print('Graph for {}: total = {}{}'.format(i,len(cat_num)))
    sns.barplot(x=cat_num, y = cat_num.index)
    plt.show()
```

```python
df['job_simp'] = df['Job Title'].apply(lambda x: 'Data Scientist' if 'data scientist' in x.lower() else
                                      'Data Analyst' if 'analyst' in x.lower()
                                       else 'Data Engineer' if 'data engineer' in x.lower()
                                      else 'Director' if 'director' in x.lower()
                                      else 'Manager' if 'manager' in x.lower()
                                      else 'ML' if 'machine learning' in x.lower()
                                      else 'na')
```

```python
df['job_simp']
```

```python
pd.pivot_table(df, index = ['job_simp','seniority'], values = 'avg_salary')
```
